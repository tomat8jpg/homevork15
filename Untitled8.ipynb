```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier, NearestCentroid
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from IPython.display import Image
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar15/BRCA_pam50.tsv', index_col = 0, sep = '\t')
df
```

```python
X = df.iloc[:, :-1].to_numpy()
y = df["Subtype"].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=17)
```

```python
model = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", KNeighborsClassifier(n_neighbors=1, p=2))
])
#аж кошки скребут что один сосед, эвклидова метрика конечно, но все же
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(accuracy_score(y_pred, y_test))
```

```python
for a in np.unique(y_test).tolist(): #делаем список из всех значений у_тест, но т.к. нампай array, делаем ту лист
    print(a,
        sum(y_pred[np.where(y_test == a)] == y_test[np.where(y_test == a)]) / len(y_test[np.where(y_test == a)])
    )
#нужно посчитать количество правильных значений и поделить на количество значений этого типа в у_тест
#находим все позиции, где в у_pred стоит вот нужный нам тип y_pred[np.where(y_test == a)
#сравниваем с игрик_тестом, таким образом вычисляя, какие позиции совпали для данного класса в предикте и в тесте
#получаем матрицу 1 или 0, затем сумма единичек - количество правильных предсказаний
#делим на одщее количество значений в у_тесте
```

классы очень сильно отличаются точностью из-за из размера: например нормал лайк самый маленький - точность нулевая, тот же люминал а огромный - точность выше средней
